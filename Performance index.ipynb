```python
table = np.array([[369,210,68],[397,73,193]])
data = np.zeros((table[0,0]+table[1,0],3))
data[:table[0,0],0]=1
data[table[0,0]:,0]=2
data[:table[0,1],1]=1
data[:table[0,2],2]=1
data[table[0,0]:table[0,0]+table[1,1],1]=1
data[table[0,0]:table[0,0]+table[1,2],2]=1

```

```python
def perf_index(data):
    oct_oct = np.sum(np.where((data[:,0]==1) & (data[:,1]==1),1,0))
    oct_mch = np.sum(np.where((data[:,0]==1) & (data[:,2]==1),1,0))
    mch_oct = np.sum(np.where((data[:,0]==2) & (data[:,1]==1),1,0))
    mch_mch = np.sum(np.where((data[:,0]==2) & (data[:,2]==1),1,0))
    oct_total = np.sum(np.where((data[:,0]==1), 1, 0))
    mch_total = np.sum(np.where((data[:,0]==1), 1, 0))    
    
    pi_oct = (oct_oct - oct_mch)/(oct_total)
    #pi_oct = (oct_oct - oct_mch)/(oct_oct+oct_mch)
    pi_mch = (mch_mch - mch_oct)/(mch_total)
    #pi_mch = (mch_mch - mch_oct)/(mch_oct+mch_mch)
    
    return (pi_oct + pi_mch)/2
print(perf_index(data))
bootstrap.ci(data, perf_index)
```
